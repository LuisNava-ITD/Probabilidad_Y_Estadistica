```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from google.colab import files


uploaded = files.upload()
df = pd.read_csv("autos2.csv")

print("Primeras filas del dataset:")
print(df.head())

print("\nEstadísticas descriptivas básicas:")
print(df.describe())

precio_min = df['price'].min()
precio_max = df['price'].max()
precio_moda = df['price'].mode()[0]
precio_media = df['price'].mean()
precio_mediana = df['price'].median()
precio_std = df['price'].std()


print("\nMedidas de tendencia central de precios:")
print(f"Mínimo: {precio_min}")
print(f"Máximo: {precio_max}")
print(f"Moda: {precio_moda}")
print(f"Media: {precio_media}")
print(f"Mediana: {precio_mediana}")
print(f"Desviación estándar: {precio_std}")


num_bins = math.ceil(1 + 3.322 * math.log10(len(df)))  # Regla de Sturges
plt.figure(figsize=(10, 5))
plt.hist(df['price'], bins=num_bins, edgecolor='black', alpha=0.7)
plt.xlabel('Precio del automóvil')
plt.ylabel('Frecuencia')
plt.title('Histograma de precios de automóviles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


plt.figure(figsize=(10, 5))
sns.boxplot(x='body-style', y='price', data=df)
plt.xlabel('Tipo de vehículo')
plt.ylabel('Precio del automóvil')
plt.title('Diagrama de caja: Precio vs Tipo de Vehículo')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.xlabel('Tracción del vehículo')
plt.ylabel('Precio del automóvil')
plt.title('Diagrama de caja: Precio vs Tracción del Vehículo')
plt.show()


plt.figure(figsize=(10, 5))
sns.boxplot(x='engine-location', y='price', data=df)
plt.xlabel('Ubicación del motor')
plt.ylabel('Precio del automóvil')
plt.title('Diagrama de caja: Precio vs Ubicación del Motor')
plt.show()

```
